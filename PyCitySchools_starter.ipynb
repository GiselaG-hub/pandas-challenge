```python
# Dependencies and Setup
import pandas as pd
from IPython.display import display

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
school_data_complete
```

```python
#Calculate the total number of schools
total_number_schools_df = len(school_data_complete["School ID"].unique())
total_number_schools_df
```

```python
#Calculate the total number of students
total_number_students_df = len(school_data_complete["Student ID"].unique())
total_number_students_df
```

```python
#Calculate the total budget
total_budget_df = sum(school_data_complete["budget"].unique())
total_budget_df
```

```python
#Calculate the average math score
average_math_score_df = school_data_complete["math_score"].mean()
average_math_score_df
```

```python
#Calculate the average reading score
average_reading_score_df = school_data_complete["reading_score"].mean()
average_reading_score_df
```

```python
#Calculate the percentage of students with a passing math score (70 or greater)
pass_math_df = school_data_complete.query('math_score >= 70')['School ID'].count()/ total_number_students_df * 100
pass_math_df
```

```python
pass_reading_df = school_data_complete.query('reading_score >= 70')['School ID'].count()/ total_number_students_df * 100
pass_reading_df
```

```python
#Calculate the percentage of students who passed math and reading (% Overall Passing)
overall_passing = school_data_complete[(school_data_complete['math_score'] >= 70) & (school_data_complete['reading_score'] >= 70)]['student_name'].count()/total_number_students_df*100
overall_passing
```

```python
#Create a dataframe to hold the above results
District_summary = pd.DataFrame({"Total Schools": [total_number_schools_df],
                        "Total Students": [total_number_students_df],
                        "Total Budget": [total_budget_df],
                       "Average Math Score": [round(average_math_score_df,2)],
                       "Average Reading Score": [round(average_reading_score_df,2)],
                       "% Passing Math": [round(pass_math_df,2)],
                       "% Passing Reading": [round(pass_reading_df,2)],
                       "Overall Passing": [overall_passing]})
District_summary
```

```python
#Optional: give the displayed data cleaner formatting
# List all the columns in the table

District_summary_table = pd.DataFrame(District_summary)

District_summary_table ["Total Students"] = District_summary_table ["Total Students"].map("{:,}".format)
District_summary_table ["Total Budget"] = District_summary_table ["Total Budget"].map("${:,.2f}".format)
District_summary_table ["Average Math Score"] = District_summary_table ["Average Math Score"].map("{:,.2f}%".format)
District_summary_table ["Average Reading Score"] = District_summary_table ["Average Reading Score"].map("{:,.2f}%".format)
District_summary_table ["% Passing Math"] = District_summary_table ["% Passing Math"].map("{:,.2f}%".format)
District_summary_table ["% Passing Reading"] = District_summary_table ["% Passing Reading"].map("{:,.2f}%".format)
District_summary_table ["Overall Passing"] = District_summary_table ["Overall Passing"].map("{:,.2f}%".format)

District_summary_table
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:

```python
#School data complete

school_data_complete["pass_reading_df"] = school_data_complete["reading_score"] >= 70
school_data_complete["pass_math_df"] = school_data_complete["math_score"] >= 70


school_data_complete
```

```python
# Group by school name

school_name = school_data_complete.groupby(["school_name"]).mean()

school_name ["Per Student Budget"] = school_name["budget"]/school_name["size"]
school_name ["% Passing Math"] = round(school_name["pass_math_df"]*100,2)
school_name ["% Passing Reading"] = round(school_name["pass_reading_df"]*100,2)
school_name ["% Overall Passing"] = ((average_math_score_df + average_reading_score_df)/2)

#overall_passing["% Overall Passing"]  = school_name[(school_name['math_score'] >= 70) & (school_name['reading_score'] >= 70)]['student_name'].count()/total_number_students_df*100
#overall_passing

school_name
```

```python
# Merge to collect information

school_data_summary = pd.merge(school_name, school_data, how="left", on=["school_name", "school_name"])
del school_data_summary['size_y']
del school_data_summary['budget_y']
del school_data_summary['Student ID']
del school_data_summary['School ID_x']

school_data_summary
```

```python
#Create a dataframe to store the results

school_summary_dataframe = pd.DataFrame({"School Name":  school_data_summary["school_name"],
                                "School Type": school_data_summary["type"],
                               "Total Students":school_data_summary["size_x"],
                               "Total School Budget": school_data_summary["budget_x"],
                               "Per Student Budget":school_data_summary["Per Student Budget"], 
                               "Average Math Score":round(school_data_summary["math_score"],2),
                               "Average Reading Score":round(school_data_summary["reading_score"],2), 
                               "% Passing Math": school_data_summary["% Passing Math"],
                               "% Passing Reading": school_data_summary["% Passing Reading"],
                               "% Overall Passing Rate": school_data_summary["% Overall Passing"]}) 


school_summary_dataframe["Total Students"] = school_summary_dataframe["Total Students"].map("{:,.0f}".format)
school_summary_dataframe["Total School Budget"] = school_summary_dataframe["Total School Budget"].map("${:,.2f}".format)
school_summary_dataframe["Per Student Budget"] = school_summary_dataframe["Per Student Budget"].map("${:,.2f}".format)

school_summary_dataframe
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
# Create a table that lists the average Reading Score for students of each grade level 
# (9th, 10th, 11th, 12th) at each school.

nineth_grade=  school_data_complete[school_data_complete["grade"] == "9th"].groupby("school_name").mean()["math_score"]
tenth_grade =  school_data_complete[school_data_complete["grade"] == "10th"].groupby("school_name").mean()["math_score"]
eleventh_grade =  school_data_complete[school_data_complete["grade"] == "11th"].groupby("school_name").mean()["math_score"]
twelveth_grade=  school_data_complete[school_data_complete["grade"] == "12th"].groupby("school_name").mean()["math_score"]

#Combine the series 
math_grade_dataframe = pd.DataFrame({"9th": nineth_grade, "10th":tenth_grade, 
                                     "11th":eleventh_grade, "12th":twelveth_grade})            

#Displayed data cleaner formatting

math_grade_dataframe[["9th","10th","11th","12th"]] = math_grade_dataframe[["9th","10th","11th","12th"]].applymap("{:.2f}".format)

math_grade_dataframe
```
